```python
import numpy as np

X = np.array([[1, 1], [1, 2], [1, 3]])
Y = np.array([[2], [4], [5]])

print("Input Matrix X:")
print(X)
print("\nShape of X:", X.shape)

print("\nTarget Vector Y:")
print(Y)
print("\nShape of Y:", Y.shape)



np.random.seed(42) # for reproducibility
w = np.random.rand(2, 1)
b = np.random.rand(1)

print("Initial Weight Vector w:")
print(w)
print("Shape of w:", w.shape)

print("\nInitial Bias b:")
print(b)
print("Shape of b:", b.shape)

```

```python

learning_rate = 0.1
n_epochs = 30

training_errors = []

for epoch in range(n_epochs):
    # Calculate predictions
    Y_pred = X @ w + b

    # Calculate error
    error = Y_pred - Y

    # Calculate Mean Squared Error (MSE)
    mse = np.mean(error**2)
    training_errors.append(mse)

    # Calculate gradients
    dw = (X.T @ error) / len(X)
    db = np.sum(error) / len(X)

    # Update weights
    w = w - learning_rate * dw
    b = b - learning_rate * db

print("\nFinal trained Weight Vector w:")
print(w)
print("\nFinal Bias b:")
print(b)
print("\nTraining Errors (MSE) per epoch:")
for i, err in enumerate(training_errors):
    print(f"Epoch {i+1}: {err:.4f}")

```

```python
Y_pred_final = X @ w + b

mse_final = np.mean((Y_pred_final - Y)**2)
rmse_final = np.sqrt(mse_final)
mae_final = np.mean(np.abs(Y_pred_final - Y))

print(f"\nFinal Mean Squared Error (MSE): {mse_final:.4f}")
print(f"Final Root Mean Squared Error (RMSE): {rmse_final:.4f}")
print(f"Final Mean Absolute Error (MAE): {mae_final:.4f}")





import matplotlib.pyplot as plt

# Assuming X[:, 1] is the feature we are plotting against Y
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 1], Y, label='Original Data')
plt.plot(X[:, 1], Y_pred_final, color='red', label='Fitted Line')
plt.xlabel('Feature (X)')
plt.ylabel('Target (Y)')
plt.title('Linear Regression Fit')
plt.legend()
plt.grid(True)
plt.show()
```
