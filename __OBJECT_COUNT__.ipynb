```python
!pip install ultralytics

#object count
# Step 1: Imports
import cv2
from ultralytics import YOLO
from google.colab import files

# Step 2: Upload video file
uploaded = files.upload()
video_path = list(uploaded.keys())[0]   # get uploaded filename

# Step 3: Load YOLO model (nano version for speed)
model = YOLO("yolov8n.pt")

# Step 4: Open video
cap = cv2.VideoCapture(video_path)
assert cap.isOpened(), "Error reading video file"

# Step 5: Video writer setup (output video)
w, h, fps = (int(cap.get(x)) for x in (cv2.CAP_PROP_FRAME_WIDTH,
                                       cv2.CAP_PROP_FRAME_HEIGHT,
                                       cv2.CAP_PROP_FPS))
out = cv2.VideoWriter("output_counting.mp4",
                      cv2.VideoWriter_fourcc(*"mp4v"), fps, (w, h))

# Step 6: Define counting line (horizontal line example)
line_y = h // 2   # middle of the frame
object_counter = 0

# Step 7: Process video frame by frame
while cap.isOpened():
    success, frame = cap.read()
    if not success:
        print("Video processing complete.")
        break

    # Run YOLO detection
    results = model(frame)

    # Get detections (bounding boxes)
    for box in results[0].boxes:
        x1, y1, x2, y2 = box.xyxy[0].cpu().numpy()  # bounding box coords
        cx = int((x1 + x2) / 2)  # center x
        cy = int((y1 + y2) / 2)  # center y

        # Draw bounding box
        cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0,255,0), 2)
        cv2.circle(frame, (cx, cy), 4, (0,0,255), -1)

        # Check if object crosses the line
        if abs(cy - line_y) < 5:   # tolerance of 5 pixels
            object_counter += 1

    # Draw counting line
    cv2.line(frame, (0, line_y), (w, line_y), (255,0,0), 2)

    # Display counter
    cv2.putText(frame, f'Count: {object_counter}', (50,50),
                cv2.FONT_HERSHEY_SIMPLEX, 1, (0,0,255), 2)

    # Write annotated frame to output video
    out.write(frame)

# Step 8: Release resources
cap.release()
out.release()
cv2.destroyAllWindows()

# Step 9: Download the output video (Colab)
files.download("output_counting.mp4")
```
